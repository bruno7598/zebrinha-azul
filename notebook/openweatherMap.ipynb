```python
import requests
import json
from datetime import datetime

api_key = '8e4a3c2c7638bbd6d1a2477967c208fe'

capitais = {
    'São Paulo': {'lat': -23.55052, 'lon': -46.633308},
    'Rio de Janeiro': {'lat': -22.906847, 'lon': -43.172896},
    'Belo Horizonte': {'lat': -19.916681, 'lon': -43.934493},
    'Porto Alegre': {'lat': -30.034647, 'lon': -51.217658},
    'Curitiba': {'lat': -25.428954, 'lon': -49.267137},
    'Salvador': {'lat': -12.9714, 'lon': -38.5014},
    'Fortaleza': {'lat': -3.71722, 'lon': -38.5434},
    'Recife': {'lat': -8.0476, 'lon': -34.8770},
    'Manaus': {'lat': -3.1190, 'lon': -60.0217},
    'Brasília': {'lat': -15.8267, 'lon': -47.9218},
}

# URL base da API
url = "https://api.openweathermap.org/data/2.5/weather"
params = {
        'lat': -15.8267,
        'lon': -47.9218,
        'appid': api_key,
        'units': 'metric', 
        'lang': 'pt'
    }

def obter_clima(cidade, coords):
    params = {
        'lat': coords['lat'],
        'lon': coords['lon'],
        'appid': api_key,
        'units': 'metric', 
        'lang': 'pt'
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return f"Erro na solicitação: {response.status_code} - {response.text}"

info_climatica_por_estado = {}

# Obter o clima para todas as capitais
for cidade, coords in capitais.items():
    climas = obter_clima(cidade, coords)
    info_climatica = {
        "Coordenadas Geográficas": {
            "Latitude": coords['lat'],
            "Longitude": coords['lon']
        },
        "Condições Climáticas Atuais": {
            "Descrição": climas['weather'][0]['description'],
            "Temperatura": climas['main']['temp'],
            "Sensação Térmica": climas['main']['feels_like'],
            "Pressão Atmosférica": climas['main']['pressure'],
            "Umidade": climas['main']['humidity'],
            "Velocidade do Vento": climas['wind']['speed'],
            "Direção do Vento": climas['wind']['deg']
        }
    }
    info_climatica_por_estado[cidade] = info_climatica

print(json.dumps(info_climatica_por_estado, indent=4, ensure_ascii=False))
```
